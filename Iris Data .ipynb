```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Iris (1).csv")
```

```python
df
```

```python
x = df.iloc[:,1:5].values
```

```python
y = df.iloc[:,5].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.2 , random_state = 0)
```

```python
#KNN Classifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
from sklearn.model_selection import cross_val_score
```

```python
Knn_Classifier = KNeighborsClassifier(n_neighbors=3, metric = "euclidean")
```

```python
Knn_Classifier.fit(x_train, y_train)
                                    
```

```python
y_pred = Knn_Classifier.predict(x_test)
```

```python
y
```

```python
y_pred
```

```python
#prediction of category which it belongs to
```

```python
Knn_Classifier.predict([[5.2,3.4,0.7, 4.5]])
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Accuracy Using KNN Classifier
```

```python
accuracy_knn = accuracy_score(y_test,y_pred)*100
```

```python
print(str(round(accuracy_knn, 2)) + "%")
```

```python
# Graph representing the Best Fitline
```

```python
plt.scatter (y_test, y_pred)
plt.plot (y_test,y_pred)
plt.show()
```
